```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Generate two vectors with 15 random floats each and plot them
vectorA = np.random.rand(15)
vectorB = np.random.rand(15)

plt.plot(range(1,16),vectorA,marker='o',color='red',label='Vector A')
plt.plot(range(1,16),vectorB,marker='o',color='blue',label='Vector B')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Graph of Vector A and Vector B')
plt.legend()
plt.grid(True)
plt.show()


```

```python
# Create a 4*4 matrix with random values and visualize as heatmap

matrix1 = np.random.rand(4, 4)
sns.heatmap(matrix1, annot=True, cmap='Blues', fmt='.2f', linewidths=.5)
plt.title('Heatmap of Matrix1')
plt.xlabel('Column')
plt.ylabel('Row')
plt.show()
```

```python
# 4*4 matrixes perform add,sub,mul and visualize bar plot


X = np.random.randint(1,10,(4,4))
Y = np.random.randint(1,10,(4,4))

print("Matrix X:\n",X)
print("\nMatrix Y:\n",Y)

matrix_sum = X + Y
print("\nMatrix Addition:\n",matrix_sum)

matrix_diff = X - Y
print("\nMatrix Difference: \n",matrix_diff)

matrix_mul = np.dot(X,Y)
print("\nMatrix Multiplication:\n",matrix_mul)

# Create the subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot Matrix Addition
axs[0, 0].bar(range(1, 17), matrix_sum.flatten())
axs[0, 0].set_title('Matrix Addition')
axs[0, 0].set_xlabel('Index')
axs[0, 0].set_ylabel('Value')

# Plot Matrix Difference
axs[0, 1].bar(range(1, 17), matrix_diff.flatten())
axs[0, 1].set_title('Matrix Difference')
axs[0, 1].set_xlabel('Index')
axs[0, 1].set_ylabel('Value')

# Plot Matrix Multiplication
axs[1, 0].bar(range(1, 17), matrix_mul.flatten())
axs[1, 0].set_title('Matrix Multiplication')
axs[1, 0].set_xlabel('Index')
axs[1, 0].set_ylabel('Value')

# Hide the last unused subplot (bottom-right corner)
axs[1, 1].axis('off')

# Adjust layout for better readability
plt.tight_layout()
plt.show()
```
